```python
# -*- coding: utf-8 -*-
import os
import pandas as pd 
import numpy as np

from sklearn import svm
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import GridSearchCV
```

```python
os.chdir("makecsv/")
df = pd.read_csv("onehotencorder.csv",header = 0)
```

```python
df.head()
```

```python
data = df[list(df.columns)[1:]]
data.head()
```

```python
answers = df['successorfail'] 

X_train, X_test, y_train, y_test = train_test_split(data, answers, train_size=0.8, test_size=0.2, random_state=1)
```

```python
X_train.count()
```

```python
X_test.count()
```

```python
lr = linear_model.LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_predict = lr.predict(X_test)
```

```python
train_score = mean_squared_error(y_test,y_predict)
```

```python
train_score
```

```python
print("トレーニングデータに対する正解率：" + str(train_score * 100) + "%")
```
